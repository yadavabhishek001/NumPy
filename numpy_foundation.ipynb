```python
data = {"Name" : ["Alice", "Bob", "Charlie"], "Age" : [25, 30, 35], "Marks" : [85, 90, 78]}
print(data)
type(data)
```

```python
# pandas automatically adds a default index: This is called a RangeIndex(start from 0 by default)
marks = pd.DataFrame(data)
print(marks)
```

```python
# we can add a custom index to the DataFrame
marks = pd.DataFrame(data, index=["Student1", "Student2", "Student3"])
print(marks)
```

```python
marks.set_index("Name", inplace=True)
 # if we are setting inplace = true it means we are changing the original object itself .if we do not set inplace = true then we have to assign the result to a new variable or to the same variable to see the changesprint(marks)
```
