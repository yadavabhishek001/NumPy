```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sammer"],
               "Age": [22, 23, None, 22, 25, 24, 26] ,
                "Marks": [85, None, 88, 90, 56, 72, None],
                "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]})
```

```python
df.shape
```

```python
#display datatypes of each columns
df.dtypes
```

```python
#generate summary statistics of the dataframe
df.describe()
```

```python
#Filter and group the data
#Create a view showing only the 'Name' and 'Marks' columns
df[["Name", "Marks"]]
```

```python
#Filtering : Select rows where 'Marks' is greater than 80
df[df["Marks"] > 80]
```

```python
# Complex filtering : Select students older than 22 who scored more than 80 in 'Math'
df[(df["Age"] > 22) & (df["Marks"] > 80) & (df["Subject"] == "Math")]
```

```python
#Aggregation : count how many students are enrolled in each subject
df.groupby("Subject").size()
```

```python
#Task : Order the data to find rankers

#Sort the dataframe by 'Marks' in descending order
df.sort_values(by="Marks", ascending=False)
```

```python
#Use a specific Pandas function to find the top 2 students based on their marks without sorting the entire dataframe manually
df.nlargest(2, "Marks")
```

```python
#Identify the student with the lowest marks(Remedial Candidate)
df.loc[df["Marks"].idxmin()]
```

```python
#Handle missing values : Fill missing 'Marks' with the average marks

# Calculate the total number of the missing values in the each column
df.isnull().sum()
```

```python
#Filter and display rows where 'Marks' are missing
df[df['Marks'].isnull()]
```

```python
# Create a new dataFrame alled cleaned_df by removing any row that conatains even a single missing value
cleaned_df = df.dropna()
```

```python
# Compare the shape of cleaned_df vs  orifinal df
print("Original DataFrame shape:", df.shape)
print("Cleaned DataFrame shape:", cleaned_df.shape)
```
