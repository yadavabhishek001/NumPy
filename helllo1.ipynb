```python
import pandas as pd
```

```python
data = {"Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sammer"],
               "Age": [22, 23, None, 22, 25, 24, 26] ,
                "Marks": [85, None, 88, 90, 56, 72, None],
                "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]}

df = pd.DataFrame(data)
print(df)
```

```python
#basic groupby
df.groupby("Subject")["Marks"].mean() #mean of marks per subject
```

```python
df.groupby("Subject")['Name'].count()
```

```python
df.groupby("Subject").size() #count of students per subject
```

```python
#Multiple aggregation
df.groupby("Subject")["Marks"].agg(["mean", "min", "max", "std"])
```

```python
#Groupby with multiple columns
df.groupby(["Subject", "Age"])["Marks"].mean()
```
